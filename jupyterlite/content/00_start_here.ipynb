## Smoke test

```
from holonic import HolonicDataset

ds = HolonicDataset()
ds.add_holon("urn:holon:demo", "Demo Holon")
ds.add_interior("urn:holon:demo", '''
    @prefix schema: <https://schema.org/> .
    <urn:item:1> a schema:Thing ; schema:name "Hello from the browser" .
''')
print(ds.summary())
```

### What works here vs. what doesn't

Notebooks 01-12 run in the browser. Notebook 13 requires `holonic[viz]` and a local Jupyter install because yFiles widgets need a server extension that Pyodide can't provide.

Also not available in-browser: `FusekiBackend` (no HTTP in the sandbox) and `holonic-migrate-registry` CLI (no subprocess invocation).
