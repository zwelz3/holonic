# holonic — Example Notebooks

Interactive examples for the [holonic](https://pypi.org/project/holonic/) library.
Each notebook is self-contained and can be run independently.
Run them in order the first time; after that, jump to whichever
topic you need.

> **No local install?**  These notebooks also run in the browser
> via [JupyterLite](https://zwelz3.github.io/holonic/jupyterlite/index.html)
> (notebook 13 excluded — it requires yFiles server extensions).

---

## Foundations

The structural primitives: holons, portals, and projections.

| # | Notebook | What you'll learn |
|---|----------|-------------------|
| 01 | [Holon Basics](01_holon_basics.ipynb) | Create holons, add multi-graph interiors, declare SHACL boundary shapes, validate the membrane |
| 02 | [Portal Traversal](02_portal_traversal.ipynb) | Register portals between holons, discover paths via SPARQL, traverse with CONSTRUCT, record PROV-O provenance |
| 03 | [Projections](03_projections.ipynb) | Type collapse, literal collapse, blank-node folding, projection pipelines, holarchy-level projection |

After these three you understand the four-layer model, how data
crosses boundaries, and how projections reshape interiors for
external consumption.

## Governance & Integration

Governing the boundaries: who owns what, which ontology is
normative, and how cross-vocabulary translation works.

| # | Notebook | What you'll learn |
|---|----------|-------------------|
| 04 | [Cross-Vocabulary Translation](04_cco_to_schemaorg.ipynb) | Translate CCO person records to Schema.org via a portal CONSTRUCT query |
| 05 | [Governed Boundary Contracts](05_governed_boundaries.ipynb) | Alignment holons, data stewardship, classification levels, `PortalClassificationShape`, full governance-chain SPARQL |
| 06 | [Holon Subtypes & Classification](06_holon_subtypes.ipynb) | AgentHolon, AggregateHolon shapes, `ClassificationLevel` enum, SHACL validation per subtype |

Notebook 04 shows the translation mechanics; notebook 05 layers
governance metadata on top (who authorized the translation, under
what standard, at what classification level). Notebook 06 covers
the functional subtype taxonomy and the shapes that enforce
per-subtype structural expectations.
